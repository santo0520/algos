```python
class Max_weight_independent_set_path_grapgh(): 
    def __init__(self,node_list): 
        node_list.insert(0,0)
        self.node_list = node_list
        #initiate max_val list to hold best val for each subgraph , right shifted 
        self.max_val = [0,self.node_list[1]]
        #run dynamic algo to fill max_val 
        for i in range(2,len(node_list),1): 
            self.max_val.append(max(self.max_val[i-1],self.max_val[i-2]+self.node_list[i]))
        
        
    
    def get_max_val(self,size):
        """return the max val of MIS for subgraph"""
        return self.max_val[size]
    def get_maximun_weigt_set(self,size):
        mis = []
        node = len(self.node_list)-1
        while node > 0: 
#             print(node)
            if self.max_val[node-1] < self.max_val[node-2] + self.node_list[node]: 
                mis.insert(0,self.node_list[node])
                node -= 2 
            else:
                node -= 1 
        return mis
            

nodes = [1,4,5,4] #should return 8 
solver = Max_weight_independent_set_path_grapgh(nodes)
print(solver.max_val)
print(solver.get_maximun_weigt_set(4))
```

```python
class Sequence_alignment():
    def __init__(self,seq1,seq2,scoring):
        """scoring should contain the score for every letter combination as dict"""
        import numpy as np
        self.scoring = scoring
        self.seq1 = seq1
        self.seq2 = seq2
        self.penalty = np.zeros((len(seq1)+1,len(seq2)+1),dtype = int)
        self.penalty[0,:] = [i*scoring["gap"] for i in range(self.penalty.shape[1])]
        self.penalty[:,0] = [i*scoring["gap"] for i in range(self.penalty.shape[0])]
        for i in range(1,self.penalty.shape[0],1):
            for j in range(1,self.penalty.shape[1],1): 
                #decide which case to fill 
                case1 = self.penalty[i-1,j-1] + self.score(seq1[i-1],seq2[j-1])
                case2 = self.penalty[i-1,j] + scoring["gap"]
                case3 = self.penalty[i,j-1] + scoring["gap"]
                self.penalty[i,j] = max([case1,case2,case3])
                
    def score(self,letter1,letter2):
        return self.scoring["match"] if letter1 == letter2 else self.scoring["mismatch"]
        
        
    def get_score(self,seq1_len,seq2_len):
        return self.penalty[seq1_len,seq2_len]
    
    def get_solution(self,seq1_len,seq2_len):
        align1 = []
        align2 = []
        match = []
        current_i,current_j = seq1_len,seq2_len
        while current_i != 0 and current_j != 0:
            #decide which case was used 
            score = self.penalty[current_i,current_j]
            if (self.penalty[current_i-1,current_j-1] + self.score(self.seq1[current_i-1],self.seq2[current_j-1])) == score:#case 1 
                align1.insert(0,self.seq1[current_i-1])
                align2.insert(0,self.seq2[current_j-1])
                match.insert(0,"|" if self.seq1[current_i-1] == self.seq2[current_j-1] else " ")
                current_i -= 1
                current_j -= 1 
            elif (self.penalty[current_i-1,current_j] + self.scoring["gap"])  == score: #case2
                align1.insert(0,self.seq1[current_i-1])
                align2.insert(0,"-")
                match.insert(0," ")
                current_i -=1 
            elif (self.penalty[current_i,current_j-1] + self.scoring["gap"]) == score: #case3
                align1.insert(0,"-")
                align2.insert(0,self.seq2[current_j-1])
                match.insert(0," ")
                current_j -=1
        #pad rest of string with gaps 
        if current_i == 0: 
            for _ in range(current_j):
                align1.insert(0,"-")
        else:
            for _ in range(current_i):
                align2.insert(0,"-")
        return (align1,align2,match)
        

scoring = {}
scoring["gap"] = -2 
scoring["mismatch"] = -1 
scoring["match"] = 2 
seq1 = "G A A T T C A G T T A".split(" ")
seq2 = "G G A T C G A".split(" ")

solver = Sequence_alignment(seq1,seq2,scoring)
print(solver.get_score(len(seq1),len(seq2)))
align1,align2,match= solver.get_solution(len(seq1),len(seq2))
print(align1)
print(match)
print(align2)
```
